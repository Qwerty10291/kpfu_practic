```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('data.csv', sep=',')
df.head()
```

```python
df['algeom'].value_counts().plot(kind='bar', title='Статистика оценок по алгебре и геометрии', xlabel='Оценка', ylabel='Количество студентов')
```

```python
df['matan'].value_counts().plot(kind='bar', title='Статистика оценок по математическому анализу', xlabel='Оценка', ylabel='Количество студентов')
```

```python
df['c++'].value_counts().plot(kind='bar', title='Статистика оценок по с++', xlabel='Оценка', ylabel='Количество студентов')
```

```python
df['english'].value_counts().plot(kind='bar', title='Статистика оценок по английскому', xlabel='Оценка', ylabel='Количество студентов')
```

```python
df.pivot_table(values=['algeom', 'matan', 'c++', 'english'], index='sex', aggfunc ='mean').plot(kind='bar', title='Средние оценки по предметам ', ylabel='Средняя оценка')
```
